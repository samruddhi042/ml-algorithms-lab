2. Implement Reinforcement learning model for Tic Tac Toe.

```python
import random

# Q-Learning Parameters
alpha = 0.1       # Learning rate
gamma = 0.9       # Discount factor
epsilon = 0.3     # Exploration rate
episodes = 30000  # Training episodes

Q = {}            # Q-table
```

```python
def get_state(board):
    return tuple(board)

def available_actions(board):
    return [i for i in range(9) if board[i] == ' ']

def check_winner(board):
    wins = [
        (0,1,2),(3,4,5),(6,7,8),
        (0,3,6),(1,4,7),(2,5,8),
        (0,4,8),(2,4,6)
    ]

    for a, b, c in wins:
        if board[a] == board[b] == board[c] and board[a] != ' ':
            return board[a]

    if ' ' not in board:
        return 'Draw'

    return None
```

```python
def choose_action(state, actions):
    if random.uniform(0,1) < epsilon:
        return random.choice(actions)

    q_values = [Q.get((state, a), 0) for a in actions]
    max_q = max(q_values)

    return actions[q_values.index(max_q)]
```

```python
for episode in range(episodes):

    board = [' '] * 9
    state = get_state(board)

    while True:
        actions = available_actions(board)
        action = choose_action(state, actions)

        # Agent move
        board[action] = 'X'
        next_state = get_state(board)

        result = check_winner(board)

        if (state, action) not in Q:
            Q[(state, action)] = 0

        # Win
        if result == 'X':
            Q[(state, action)] += alpha * (1 - Q[(state, action)])
            break

        # Draw
        if result == 'Draw':
            break

        # Opponent move (random)
        opp = random.choice(available_actions(board))
        board[opp] = 'O'

        result = check_winner(board)

        # Loss
        if result == 'O':
            Q[(state, action)] += alpha * (-1 - Q[(state, action)])
            break

        # Future reward
        future_q = 0
        next_actions = available_actions(board)

        if next_actions:
            future_q = max(Q.get((next_state, a), 0) for a in next_actions)

        Q[(state, action)] += alpha * (gamma * future_q - Q[(state, action)])

        state = get_state(board)

print("Training Completed!")
```

```python
def print_board(board):
    print()
    print(board[0], "|", board[1], "|", board[2])
    print("--+---+--")
    print(board[3], "|", board[4], "|", board[5])
    print("--+---+--")
    print(board[6], "|", board[7], "|", board[8])
    print()
```

```python
def play():
    board = [' '] * 9

    while True:
        print_board(board)

        move = int(input("Enter your move (0-8): "))

        if board[move] != ' ':
            print("Invalid move")
            continue

        # Human move
        board[move] = 'O'

        if check_winner(board):
            break

        # AI move
        state = get_state(board)
        actions = available_actions(board)

        q_vals = [Q.get((state, a), 0) for a in actions]
        ai_move = actions[q_vals.index(max(q_vals))]

        board[ai_move] = 'X'

        if check_winner(board):
            break

    print_board(board)

    result = check_winner(board)

    if result == 'X':
        print("AI Wins!")
    elif result == 'O':
        print("You Win!")
    else:
        print("Match Draw!")
```

```python
play()
```
